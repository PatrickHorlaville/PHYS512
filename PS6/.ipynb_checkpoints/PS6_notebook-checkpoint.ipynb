## Problem 1

We would like to shift an array by convolution. We first write the function to shift a specified array by a specified shift:

```python
def arr_shift(arr, shift):
    
    f = arr

    g = np.zeros(len(f))
    g[shift] = 1

    F = np.fft.rfft(f)
    G = np.fft.rfft(g)
    shifted_func = np.fft.irfft(F*G)

    return shifted_func
```

Then we use it with an example on a Gaussian function

```python
x = np.linspace(-5,5,1000)
y = np.exp(-0.5*(x)**2/0.35**2)

shift = int(len(y)/2)
y_shifted = arr_shift(y, shift)

plt.plot(y, label = 'Original Gaussian')
plt.plot(np.abs(y_shifted), label = 'Shifted Gaussian')
plt.legend(loc='upper right')
```

## Problem 2

a)

Here we write a function to take the correlation of two functions:

```python
def get_correlation_func(f, g):
    
    F = np.fft.rfft(f)
    G = np.fft.rfft(g)

    corr = np.fft.irfft(F*np.conj(G))

    return corr
```

Now we take the correlation of a Gaussian with itself:

```python
x = np.linspace(-5,5,1000)
y = np.exp(-0.5*(x)**2/0.35**2)
corr_func = get_correlation_func(y, y)

plt.plot(y, label = 'Original Gaussian')
plt.plot(np.abs(corr_func), label = 'Two correlated Gaussians')
plt.legend()
```

b)

Now we would like to take the convolution of a Gaussian with itself but shifted, and we would like to visualize how the result depends on how shifted is the second Gaussian:

```python
def shift_arr(f, shift_frames):

    delta = np.zeros(len(f))
    delta[shift_frames] = 1
    F = np.fft.fft(f)
    G = np.fft.fft(delta)
    shifted_func = np.fft.ifft(F*G)

    return shifted_func

def get_correlation_func(f, g):

    F = np.fft.fft(f)
    G = np.fft.fft(g)

    return np.fft.ifft(F*np.conj(G))

def modified_correlation(f, g, frame_shift):

    shifted_f = shift_arr(f, frame_shift)
    corr = get_correlation_func(shifted_f, g)
    
    return np.abs(corr), np.abs(shifted_f)
```

```python
nframes = 1000
x = np.linspace(-5,5,nframes)
y = np.exp(-0.5*(x)**2/0.35**2)

shift = np.linspace(0, 0.99, 100)

for i in shift:
    plt.clf()
    corr, shift = modified_correlation(y, y, int(i*nframes))
    plt.plot(corr, label = 'Correlated Gaussian')
    plt.plot(y, label = 'Original Gaussian')
    plt.plot(shift, label = 'Shifted Gaussian', ls = '--')
    plt.title('Convolution of Gaussian & Shifted Gaussian, shifted by '+str(int(i*nframes))+' frames')
    plt.legend()
    plt.savefig('frames/snap'+str(int(i*nframes))+'.png')
```

## Problem 3

We want to write a routine to take an FFT-based convolution of 2 arrays without any danger of wrapping around.

First we write a function to add zeroes at the end of an array:

```python
def pad_func(arr, npad):
    
    return np.hstack((arr, np.zeros(npad)))
```

Then we write our routine. We fill our arrays with zeroes depending on the length of each array with respect to each other. Then we simply take the convolution of the two:

```python
def conv_fft(f, g):

    npts_f = len(f)
    npts_g = len(g)

    if npts_f == npts_g:
        f_pad = pad_func(f, npts_f)
        g_pad = pad_func(g, npts_g)

    elif npts_f > npts_g:
        diff = npts_f - npts_g
        f_pad = pad_func(f, npts_f)
        g_pad = pad_func(g, npts_f + diff)

    else:
        diff = npts_g - npts_f
        f_pad = pad_func(f, npts_g + diff)
        g_pad = pad_func(g, npts_g)

    f_fft = np.fft.fft(f_pad)
    g_fft = np.fft.fft(g_pad)

    convolution = np.fft.ifft(f_fft*g_fft)

    return np.abs(convolution)
```

Let's try our routine on an exponential function and a shifting delta function:

```python
x = np.linspace(-5,5,1000)
y = np.exp(-0.5*(x)**2/0.35**2)

shifts = np.linspace(0, 1.9999, 100)

for shift in shifts:
    plt.clf()
    g = np.zeros(2*len(y))
    g[int(shift*len(y))] = 1
    conv = conv_fft(y, g)
    plt.plot(conv)
    plt.title('Convolution of Gaussian & Delta function, shifted by '+str(int(shift*len(y)))+' frames')
    plt.savefig('frames_q3/snap'+str(int(shift*len(y)))+'.png')
```

## Problem 4

c)

In the PDF attached, there's the analytic derivation of the DFT of a sine wave with a non-integer k. Here, we use that solution and try to see what it looks like. We compare the DFT of this sine wave with the real analytic FFT from numpy, and we look at the residuals between the two results.

```python
def analytic_fft(k, N):

    my_fft = np.zeros(N, dtype = 'complex')

    for i in range(0, N):
        my_fft[i] = (1/2J)*((1 - np.exp(-2J*np.pi*(i - k)))/(1 - np.exp(-2J*np.pi*(i - k)/N)) - (1 - np.exp(-2J*np.pi*(i + k)))/(1 - np.exp(-2J*np.pi*(i + k)/N)))

    return my_fft
```

We use an arbitrary value k that is not an integer:

```python
N = 50
x = np.arange(N)
k = 12.5
y = np.sin(2*np.pi*k*x/N)
```

We only look at the real part of our DFT so we truncate our array:

```python
fft_analytic = np.abs(analytic_fft(k, N))
rfft_analytic = fft_analytic[0 : N//2 + 1]
fft_numpy = np.abs(np.fft.rfft(y))
```

```python
plt.plot(fft_numpy[:], label = 'numpy rFFT')
plt.plot(rfft_analytic[:], label = 'Real Analytic FFT', ls = '--')
plt.title('Real analytic FFT versus numpy rFFT')
plt.xlabel('Modes k')
plt.ylabel('Power P in each mode k')
plt.legend()
```

We are close to a delta function but not quite!

Now, we look at the residuals:

```python
plt.plot((fft_numpy - rfft_analytic)[:])
plt.title('Residuals')
plt.xlabel('Modes k')
plt.ylabel('Power P in each mode k')
```

```python
err = np.abs(np.std((fft_numpy - rfft_analytic)[:]))
print('The margin of error between the numpy FFT and our function is of order', round(err, 14))
```

d)

Here we apply a window on our calculations to try to see if the spectral leakage decreases. We apply the window as prescribed and compare with our previous results:

```python
window = 0.5*(1 - np.cos(2*np.pi*x/N))
fft_windowed = np.abs(np.fft.rfft(y*window))

plt.plot(fft_numpy, label = 'numpy rFFT')
plt.plot(fft_windowed, label = 'Windowed numpy rFFT')
plt.legend()
plt.title('Windowed real FFT')
plt.xlabel('Modes k')
plt.ylabel('Power P in each mode k')
```

Indeed, the power relocates towards the center of the delta function. It works !

e)

We want to show that the Fourier transform of the window follows specific values (first two values are N/2 and -N/4, and last value is -N/4. The rest are 0's). Hence we first compute the FFT of the window, build the expectation of what it should look like, and plot the comparison:

```python
window = 0.5*(1 - np.cos(2*np.pi*x/N)) 
fft_window = np.fft.fft(window)
```

```python
expected = np.zeros(len(fft_window))
expected[0] = N/2
expected[1] = -N/4
expected[-1] = -N/4
```

```python
plt.plot(fft_window, label = 'FFT of window')
plt.plot(expected, label = 'Expected', ls = '--')
plt.legend()
plt.xlabel('Modes k')
plt.ylabel('Power P in each mode k')
```

Indeed, the result follows our expectations! We have the residuals:

```python
plt.plot(fft_window - expected)
plt.xlabel('Modes k')
plt.ylabel('Power P in each mode k')
plt.title('Residuals')
```

We can use those results to retrieve our windowed FFT:

```python
yft = np.fft.rfft(y)
yft_smooth = 0.5*yft - 0.25*np.roll(yft,1) - 0.25*np.roll(yft, -1)
```

```python
plt.plot(np.abs(yft_smooth), label = 'smoothed FFT')
plt.plot(fft_windowed, label = 'Windowed FFT', ls = '--')
plt.legend()
plt.xlabel('Modes k')
plt.ylabel('Power P in each mode k')
```
