## Problem 2

We would like to generate exponential deviates using the transformation of other distributions. Which of Lorentzian, Gaussian or power law could I use for my bounding distribution?

- Remember that we want our bounding distribution to always lie above our original distribution. Also, the closer to the original, the better our bounding distribution.

- So we want a distribution that lies above the exponential and that looks the most like the exponential

```python
y = np.linspace(0, np.pi/2, 1001)
y = 0.5*(y[1:] + y[:-1])
```

```python
plt.plot(y, (1/(1+y**2)), label = "Lorentzian")
plt.plot(y, np.exp(-y), label = "Exponential")
plt.plot(y, np.exp((-y**2)/2), label = "Gaussian")
plt.title('Comparing distributions')
plt.legend()
```

From visual analysis, the Lorentzian seems to be the most suited to our task as its shape looks more like the one of the exponential than the Gaussian. 

As in class, we will use the example for $x = tan(y)$.

Now, recall we have $P'(y) = P(x)\frac{dx}{dy}$. Here, $P(x) = e^{-x}$ and $\frac{dx}{dy} = \frac{d(tan(y))}{dy} = sec^2 (y)$, such that $P'(y) = sec^2 (y)e^{-tan(y)}$.

So we set:

```python
y = np.linspace(0, np.pi/2, 1001)
y = 0.5*(y[1:] + y[:-1])
p = (1/np.cos(y)**2)*np.exp(- np.tan(y))
```

```python
plt.plot(y, p)
plt.title(r"$P'(y) = sec^2 (x) e^{-tan(y)}$")
plt.xlabel('y')
plt.ylabel("P'(y)")
```

Using code snippets from 'rand_arctan.py', we proceed with computing the deviates and comparing them to the predicted exponential distribution:

```python
n = 1000000
yy = np.random.uniform(low = 0, high = np.pi/2, size = n)
myp = (1/np.cos(yy)**2)*np.exp(- np.tan(yy))
fac = 1.01*p.max()
accept = (np.random.rand(n)*fac) < myp
print('accept fraction is ', np.mean(accept))
```

For the range of our bins, we cover up to $\pi$

```python
y_use = yy[accept]
x_use = np.tan(y_use)
aa, bb = np.histogram(x_use, np.linspace(0, np.pi, 100))
b_cent = 0.5*(bb[1:] + bb[:-1])
pred = np.exp(-b_cent)
pred = pred/pred.sum()
aa = aa/aa.sum()
```

```python
plt.plot(b_cent, pred, 'k')
plt.bar(b_cent, aa, 0.1)
plt.title('Distribution of deviates with exponential prediction plotted')
```

The match is visually accurate!

## Problem 3

Here we would like to make our distribution of deviates using ratio of uniforms. We start by generating random $u$ and $v$ numbers. We pose a condition on $u$: it has to be smaller than an upper bound defined by $u < \sqrt{P(r)}$ where $r = \frac{v}{u}$ and where $P(r) = e^{-r}$. From our selection of $u$'s, we make a selection of $r$'s. We then make the exponential prediction of what the distribution of the selected $r$'s should look like based on $P(r)$. Then we plot an histogram of our $r$'s and we compare to the prediction. We use and adapt code snippets from 'ratio_uniforms_gaussian.py':

```python
N = 1000000
```

```python
random.seed(21)
u = np.random.rand(N)
random.seed(22)
v = np.random.rand(N)
r = v/u
```

```python
upper_bound_u = np.sqrt(np.exp(-r))
accept = u < upper_bound_u
u_good = u[accept]
v_good = v[accept]
expp = v_good/u_good

a, b = np.histogram(expp, 100, range = (0, 8))
bb = 0.5*(b[1:] + b[:-1])
pred = np.exp(- bb)*np.sum(accept)*(bb[2] - bb[1])
```

```python
plt.bar(bb, a, 0.15)
plt.plot(bb, pred, 'k')
plt.title('Distribution of deviates using ratio of uniforms, with exponential prediction plotted')
```

Just like in question 2, the match between the computed deviates and their prediction is pretty good visually!
